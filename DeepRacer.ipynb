```python
import math

def reward_function(params):

    MAX_REWARD = 1e2
    MIN_REWARD = 1e-3
    DIRECTION_THRESHOLD = 10


    all_wheels_on_track = params['all_wheels_on_track']
    distance_from_center = params['distance_from_center']
    track_width = params['track_width']
    steering = params['steering_angle'] # negative = turn right, positive = turn left
    is_left_of_center = params['is_left_of_center']
    speed = params['speed']
    waypoints = params['waypoints']
    closest_waypoints = params['closest_waypoints']
    heading = params['heading']

    reward = 1



    if all_wheels_on_track:
        reward *= 1.1767
    else:
        reward *= 0.85


    marker_1 = 0.1667 * track_width
    marker_2 = 0.3333 * track_width
    marker_3 = 0.5 * track_width

    if distance_from_center <= marker_1 and all_wheels_on_track:
        reward *= 1.414
    elif distance_from_center <= marker_2 and all_wheels_on_track:
        reward *= 1.289
    elif distance_from_center <= marker_3 and all_wheels_on_track:
        reward *= 1.083



    if speed >= 3:
        reward *= speed**2
    elif speed >= 2:
        reward *= speed
    elif  speed >= 1:
        reward *= speed





    next_point = waypoints[closest_waypoints[1]]
    prev_point = waypoints[closest_waypoints[0]]

    track_direction = math.atan2(next_point[1] - prev_point[1], next_point[0] - prev_point[0])
    track_direction = math.degrees(track_direction)

    direction_delta = track_direction - heading
    
    if direction_delta < 0: # should turn right
        if steering < 0:
            if steering > direction_delta:
                reward *= 2
            else:
                reward *= 1.2
        else:
            reward *= 0.7
    elif direction_delta >0: # should turn left
        if steering > 0:
            if steering < direction_delta:
                reward *= 2
            else:
                reward *= 1.2
        else:
            reward *= 0.7
    else:
        reward *= 1.5


    direction_delta_abs = abs(direction_delta)
    if direction_delta_abs > 180:
        direction_delta_abs = 360 - direction_delta_abs
    if direction_delta_abs > DIRECTION_THRESHOLD:
        reward *= 0.5
    

    return float(reward)
```
